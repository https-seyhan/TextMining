<h3>5.2 Representing <i>Description:</i> column in a vector space.</h3>

<p>Define a function for stemming words as it is not possible to inlcude stemmer as a vectorizer argument </p>
<p> Create a a series of vectorizers with slight modification for <i>Description:</i> column to examine if varying vector representation of words impacts on the model accuracy. <p>



```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re

stemmer = PorterStemmer()

#####Fucntion to stem words########

#Function to stemm words in text using Porters Stemmer
#Return only words between 2-30 characters long

def tokenize_stem(text):
    
    #Find where there is a sequence of character 2-30 lenght without break. May chop off very long word.
    tokens = re.findall(r"[a-zA-Z]{2,30}", text)
   
    #Stem the words. 
    stemmed = [] #List to hold the words
    
    #Iterate over the words in tn the text and stem.
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    
          
    return stemmed

######End of Function#############

####Create a series of methods for transform the text into a vector representation####
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer


#Values for document frequencies in percent.
minp=0.001
maxp=0.95


#1. Lower Case
vec1_count = CountVectorizer(min_df=minp,max_df=maxp)     
vec1_idf =   TfidfVectorizer(min_df=minp,max_df=maxp)

#2. Lower Case + non_letters
vec2_count = CountVectorizer(token_pattern=r"[a-zA-Z]{2,30}",min_df=minp,max_df=maxp)
vec2_idf =   TfidfVectorizer(token_pattern=r"[a-zA-Z]{2,30}",min_df=minp,max_df=maxp)

#3. Lower Case + non letter + 1-2 ngrams
vec3_count = CountVectorizer(token_pattern=r"[a-zA-Z]{2,30}",ngram_range=(1,2),analyzer='word',min_df=minp,max_df=maxp)
vec3_idf =   TfidfVectorizer(token_pattern=r"[a-zA-Z]{2,30}",ngram_range=(1,2),analyzer='word',min_df=minp,max_df=maxp)

#4. Lower Case + non_letters + 1-3 ngrams
vec4_count = CountVectorizer(token_pattern=r"[a-zA-Z]{2,30}",ngram_range=(1,3),analyzer='word',min_df=minp,max_df=maxp)
vec4_idf =   TfidfVectorizer(token_pattern=r"[a-zA-Z]{2,30}",ngram_range=(1,3),analyzer='word',min_df=minp,max_df=maxp)

#5.  Lower Case + non_letters + 1-3 ngrams + stopword = 'english'
vec5_count = CountVectorizer(token_pattern=r"[a-zA-Z]{2,30}",min_df=minp,max_df=maxp,
                             ngram_range=(1,3),analyzer='word',stop_words='english')
vec5_idf =   TfidfVectorizer(token_pattern=r"[a-zA-Z]{2,30}",min_df=minp,max_df=maxp,
                             ngram_range=(1,3),analyzer='word',stop_words='english')

#6. Lower Case + non_letters + 1-3 ngrams + Porter Stemming uses defined function)
vec6_count = CountVectorizer(tokenizer=tokenize_stem,min_df=minp,max_df=maxp,ngram_range=(1,3),analyzer='word')
vec6_idf =   TfidfVectorizer(tokenizer=tokenize_stem,min_df=minp,max_df=maxp,ngram_range=(1,3),analyzer='word')
```

<h3>5.3 Example walk through with Naive Bayes </h3>

<p>Set up the sequence of steps from: data &#8594; vector space representation of <i>Description:</i> column &#8594; splitting into test data  &#8594; modeling the training data with cross validation  &#8594; cacluating the accuracy on the test data</p>


<p><b>Defined Function</b></p>
<p>Function to return confusion matrix in a dataframe with columns & index names the same as target label names.
Relies on the <b>scikit-learn.preprocessing.LabelEncoder()</b> method.<p>




```python
def nice_confusion(cArray):
    cMatrix=pd.DataFrame(cArray)
    for c in cMatrix.columns:
        cMatrix = cMatrix.rename(columns={cMatrix.columns[c]:le.inverse_transform(cMatrix.columns[c])})
        cMatrix = cMatrix.rename(index={cMatrix.index[c]:le.inverse_transform(cMatrix.index[c])})
    return (cMatrix)
```

```python
#1# Vector Rerpresentation: Use the first vector representation provided above
vec=vec1_count

#2# Specify the model
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()

#3# Using pipelineing, the data transformations can chained together 
#and the last object is the estimator or model.

from sklearn import pipeline
pipe = pipeline.Pipeline([('vectorizer', vec),
                          ('clf', clf)])


#4# Extract the predictor and target variables.

X=df['Description:']  

#Use the label encoder to transform the column
#into useable numeric array. Label encoders are useful for extract the 
#alpha representation of a label, when the model has been completed. 

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df['Warning:'])
y=le.transform(df['Warning:'])

#5# Split the data randomly into test and training sets. 
from sklearn import cross_validation
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,y,test_size=0.4,random_state=79)

#6# Train the model using 5 fold craiss validation and calculate the scores. 
train_scores = cross_validation.cross_val_score(pipe, X_train, y_train, cv=5)
#Fit the model
pipe=pipe.fit(X_train, y_train)
#Score the model on unseen test data  data
test_score=pipe.score(X_test,y_test)

#7# Print out the accuracy scores and 
print("Training Accuracy: %0.3f (+/- %0.3f). Test Score: (%0.3f)" 
      % (train_scores.mean(), train_scores.std(),test_score))


#8# For confusion matrix
from sklearn import metrics

labels = ['yes', 'no']
y_predicted = pipe.predict(X_test)
cm = metrics.confusion_matrix(y_test, y_predicted)

#call function to print confusion matrix.
cm=nice_confusion(cm)
cm
```

<h3>5.4 Multiple models and vectors</h3> <br>

Repeat step <b>5.3</b> using for loop with zip to alter the data transformation applied to <i>Description:</i> and the learning algorithm. Capture the results in a data frame and print out top 10 results based on Training Set Accuracy. (This part could be split for parallel processing.) 

```python
#All the methods used.
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import cross_validation
from sklearn import pipeline
from sklearn import preprocessing
from sklearn import metrics

# Specify the models###
clf1 = MultinomialNB()
clf2=SGDClassifier()
#clf2 = SGDClassifier(loss="log", penalty="L2") #this is best done. 

#from sklearn import svm  - svm is crap
#clf2 = svm.SVC()
#from sklearn import tree - requires dense matrix
#clf2 = tree.DecisionTreeClassifier()

# Extract the predictor and target variables.
X=df['Description:']  

#Encoder the target variable. 
le = preprocessing.LabelEncoder()
le.fit(df['Warning:'])
y=le.transform(df['Warning:'])

# Split the data randomly into test and training sets. Set the random state for comparison.
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,y,test_size=0.4,random_state=79)

#Empty list for holding the results of the models
#Results will be appended at each iteration
results = []
i=0
for vec,vec_label in zip([vec1_count,vec2_count,vec3_count,vec4_count,vec5_count,vec6_count,
                          vec1_idf,vec2_idf,vec3_idf,vec4_idf,vec5_idf,vec6_idf],
                         ["Freq-1","Freq-2","Freq-3","Freq-4","Freq-5","Freq-6",
                          "TFIDF-1","TFIDF-2","TFIDF-3","TFIDF-4","TFIDF-5","TFIDF-6"]):
    
   
   
        
    for clf,model_label in zip([clf1, clf2], ['NB', 'SGD']):
                
        pipe = pipeline.Pipeline([('vectorizer', vec),('clf', clf)])
    
        #Must fit to get the number of features.
        pipe.fit(X_train, y_train)
        number_features=len(vec.vocabulary_)  
            
        #Train the model using 5 fold cross validation and calculate the scores. 
        train_scores = cross_validation.cross_val_score(pipe, X_train, y_train, cv=5)
                
        test_score=pipe.score(X_test,y_test)

        #User a dictionary to capture model results
        results_dict = {'3.Train Acc.':train_scores.mean(),
                        '4.Train std.':train_scores.std(),
                        '5.Test Score':test_score,
                        '1.Model':model_label,
                        '2.Vector':vec_label,
                        '# Features':number_features}

        #Append each dict item to a list
        results.append(dict(results_dict))
        
        
        
     
#Convert results dictionary to a data frame
results_df=pd.DataFrame(results)
results_df=results_df.sort('3.Train Acc.',ascending=[False])
results_df.head(n=10)
```

<h3>5.5 Feature Selection </h3>

Take the best model from the previous step, based on Test Score; <b>Naive Bayes</b><i> (Vectorization = Frequency, all lower case, minimum documents occurence = 0.001%, max document occurence = 95%)</i><b> (2785 features for NB; 16847 features for SGD ) </b> and apply techniques to reduce the feature set. Currently only <b>chi-square</b> measure is implemented in <b>scikit-learn</b>.

```python
#feature_select.KBest, feature_select.SelectPercentile for determing the number of features
from sklearn.feature_selection import SelectKBest,SelectPercentile, chi2

#Vector Rerpresentation: Use the first vector representation provided above data frame. (Highest Training Accuracy)
#Use the highest Trainingg Accuracy even though have test score results to avoid overfitting. 
vec=vec2_count
#vec=vec6_idf  #- FOR SGD Model 

# Specify the model
clf = MultinomialNB()
#clf=SGDClassifier()

# Extract the predictor and target variables.
X=df['Description:']  
16
#Use the label encoder to transform the column
#into useable numeric array. Label encoders are useful for extract the 
#alpha representation of a label, when the model has been completed. 
le = preprocessing.LabelEncoder()
le.fit(df['Warning:'])
y=le.transform(df['Warning:'])

#Split the data randomly into test and training sets. 
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,y,test_size=0.4,random_state=79)

vec.fit(X_train)
X_train=vec.transform(X_train)
X_test=vec.transform(X_test)

#Results will be appended at each iteration
results = []

for sel in [150,300,450,600]:
    
    #Feature selection based on select k number of features
    fs = SelectKBest(chi2, k=sel)
    X_train_reduced = fs.fit_transform(X_train, y_train)
    X_test_reduced = fs.transform(X_test)
    
    #number of features is now determined by feature selection
    number_features=X_train_reduced.shape[1]
    
    #Train the model using 5 fold cross validation and calculate the scores.
    train_scores = cross_validation.cross_val_score(clf, X_train_reduced, y_train, cv=5)
    
    #Fit the model
    clf.fit(X_train_reduced, y_train)
    #Score the model on unseen test data  data
    test_score=clf.score(X_test_reduced,y_test)
    
    results_dict = {'1.Train Acc.':train_scores.mean(),
                    '2.Train std.':train_scores.std(),
                    '3.Test Score':test_score,
                    '# Features':number_features}
    
    #Append each dict item to a list
    results.append(dict(results_dict))


for sel in [5,10,15,20,25]:
    
    #Feature selection based on top percent of features based on chi-square meausre
    fs = SelectPercentile(chi2, percentile=sel)
    X_train_reduced = fs.fit_transform(X_train, y_train)
    X_test_reduced = fs.transform(X_test)
    
    #number of features is now determined by feature selection
    number_features=X_train_reduced.shape[1]
    
    #Train the model using 5 fold cross validation and calculate the scores.
    train_scores = cross_validation.cross_val_score(clf, X_train_reduced, y_train, cv=5)
    
    #Fit the model
    clf.fit(X_train_reduced, y_train)
    #Score the model on unseen test data  data
    test_score=clf.score(X_test_reduced,y_test)
    
    results_dict = {'1.Train Acc.':train_scores.mean(),
                    '2.Train std.':train_scores.std(),
                    '3.Test Score':test_score,
                    '# Features':number_features}
    
    #Append each dict item to a list
    results.append(dict(results_dict))
    
#Convert results dictionary to a data frame
results_df2=pd.DataFrame(results)
results_df2
```

<h3>5.6 Visualise the Results</h3>

<p>Create a scatter plot of the variables in the decision boundary. Using only 1-ngrams, select the top 250 most predictive features based the chi-square measure<p>

<p> Plot the variables with frequency count split by label <i>Warning</i>. Variables are sized by chi-square value and coloured if they are a stop word. 

```python
#####Same as step 5.5, but using binary Vectorizer#####

#Vector Rerpresentation: Use the first vector representation provided above, only letters
vec = CountVectorizer(token_pattern=r"[a-zA-Z]{2,30}",min_df=0.001,max_df=0.95,binary=True)

# Specify the model
clf = MultinomialNB()

# Extract the predictor and target variables.
X=df['Description:']  

#Use the label encoder to transform the column
#into useable numeric array. Label encoders are useful for extracting the 
#alpha representation of a label, when the model has been completed. 
le = preprocessing.LabelEncoder()
le.fit(df['Warning:'])
y=le.transform(df['Warning:'])

#Split the data randomly into test and training sets. 
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,y,test_size=0.4,random_state=79)

vec.fit(X_train)
X_train=vec.transform(X_train)


#Feature selection based on select k number of features
fs = SelectKBest(chi2, k=250)
fs.fit(X_train,y_train)


#Return the feature names from the vectorisation
#Get the list of the top 250 as calculated by chi-squared.
chi_index = fs.get_support(indices=False)

#Get feature names and chi-square score.
feature_names = vec.get_feature_names()
chi_score=fs.scores_

#Using index select top 250
from itertools import compress
features_selected=list(compress(feature_names, chi_index))
features_chi_score=list(compress(chi_score,chi_index))

#Tranform the Vectorization to a dense matrix and subset based on the list
#of 250 selected features
X_train_dense=pd.DataFrame(X_train.todense(),columns=vec.get_feature_names())
X_train_select=X_train_dense[features_selected]

#Append the target variable,using the label encoder
X_train_select['words'] = le.inverse_transform(y_train)

#Group and sum on the Target Label, counts the number of documents a word appears in. 
grouped=X_train_select.groupby('words')
features_sum=(grouped.sum()).T
features_sum['chi_score']=features_chi_score

#First stem the stopwords as the feature words have been stemmed by the vectorizer.
from nltk.corpus import stopwords
stop = stopwords.words('english')

#Create column to indicate if word is a stop word.
features_sum['stop']='blue'
i=0
for item in features_selected:
    if item in stop:
        features_sum['stop'][i]='red'
    i=i+1

features_sum=features_sum.sort('chi_score',ascending=[False])
features_sum.head(n=5)
```
